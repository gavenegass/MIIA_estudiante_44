## 1. Obtención e inspección de archivos

Cada archivo corresponde a una de las variables antes mencionadas y se puede garantizar que todos relacionan a las personas encuestadas en el mismo orden (i.e., la primera línea de todos los archivos corresponde a datos de la primera persona encuestada, la segunda línea a la segunda persona, etc.).

Los archivos con los que se cuenta son:
* *edades.txt*: un archivo ordenado con las edades de cada persona de la población encuestada 
* *género.txt*: un archivo ordenado con el género de cada persona de la población encuestada 
* *estado_civil.txt*: un archivo ordenado con el estado civil de cada persona de la población encuestada 
* *escolaridad.txt*: un archivo ordenado con el nivel de escolaridad de cada persona de la población encuestada 
* *estrato.txt*: un archivo ordenado con el estrato socio-económico de cada persona de la población encuestada
* *region.txt*: un archivo ordenado con la región de proveniencia de cada persona de la población encuestada

**Misión 1**

Poner en variables de Python la información que está en los archivos listados arriba.

* *Versión base*: escribe línea a línea el código necesario para poner en variables de Python la información de cada uno de los archivos.

* *Milla extra*: escribe una función que sea capaz de hacer lo anterior para un archivo y úsala en un ciclo para automatizar la lectura de los diferentes archivos.

```python
# código

import os

def datos_completos (path_file):
    
    dict_complete={}    
    for file in path_file:
        
        if file.endswith(".txt"):
            path="C:/Users/Teletrabajo/Desktop/UNIVERSIDAD DE LOS ANDES/1 SEMESTRE/HERRAMIENTAS COMPUTACIONALES/Laboratorio/Lab 1/Archivos/" + file
            with open(path, 'r', encoding="utf-8") as column:
                key=file[:-4]
                datos=column.read().splitlines()
          
                try:
                    float(datos[0])
                    dat_int=list(map(float, datos))

        
                    dict_complete[key]=dat_int
                    
                except:
                    dict_complete[key]=datos
        else:
            pass

    key='id'  

    lista_id=[i for i in range (1, len(dict_complete["edad"])+1)]

    dict_complete[key]=lista_id
        
    return dict_complete
     

path_file=os.listdir("C:/Users/Teletrabajo/Desktop/UNIVERSIDAD DE LOS ANDES/1 SEMESTRE/HERRAMIENTAS COMPUTACIONALES/Laboratorio/Lab 1/Archivos/")
datos=datos_completos(path_file)
```

```python
# Código Milla extra

#VARIABLES CUANTITATIVAS.

import matplotlib.pyplot as plt

def graficos_box(datos):
    fig, ax = plt.subplots(1, 2, sharey=False, figsize =(16,5))
    ax[0].boxplot(datos['promedio'])
    ax[1].boxplot(datos['edad'])    
    ax[0].set_title("Promedio academico")
    ax[1].set_title("Edad")
 
    return plt.show()
graficos_box(datos)


def graficos_box(datos):
    fig, ax = plt.subplots(1, 2, sharey=False, figsize =(16,5))
    ax[0].hist(datos['promedio'])
    ax[1].hist(datos['edad'])    
    ax[0].set_title("Promedio academico")
    ax[1].set_title("Edad")
 
    return plt.show()
graficos_box(datos)

```

### Se observa que en las variables de Edad y Promedio Académico no cuenta con valores atipicos. La mayoria de las personas cuentan con un promedio por debajo de 3.4 y están dentro de los 2 y 4 puntos. También se evidencia que la mayoria de las personas son de 32 años o más y que tienen una disribución de los datos muy uniforme es decir que todas las edades tienen una participación muy parecidas.

```python
def unique (lista):
    unicos=[]
    
    for i in lista:
        if i not in unicos:
            unicos.append(i)
    unicos.sort(reverse=True)
    return unicos


def graficar(poblacion):

    fig, axs = plt.subplots()
    axs.hist(poblacion, color='#F2AB6D', rwidth=0.85)
    axs.set_title("Histograma")
    labels=unique(poblacion)
    
graficar(datos["region"])
graficar(datos["escolaridad"])
graficar(datos["estado_civil"])
```

### Se observa que las tres regiones más representativas son: Caribe, Andina y Pacifica, también que existen pocas personas con posgrados y también existen pocas personas son profesional y por último, aunque existe una preferencia en la personas casadas, también las personas divorciadas y solteras tiene preponderancia.
