### Ejercicio 1
Utiliza `plotly` para realizar un  diagramas de torta sobre el estado conyugal de los clientes (resaltando el más predominante).

```python
# YOUR CODE HERE
df_status = pd.DataFrame(df.groupby('Marital_Status')['CLIENTNUM'].count())
df_status = df_status.sort_values(by=['CLIENTNUM'],ascending=False)

fig = go.Figure(data=[go.Pie(labels=df_status.index, values=df_status.CLIENTNUM, textinfo='label+percent',
                             insidetextorientation='radial', pull=[0.05, 0, 0, 0]
                            )],
               layout = go.Layout(
                                    title = "Distribución Estado Conyugal"
                                    
                                  )
               )
fig.show()
```

### Ejercicio 2
Utiliza `plotly` para realizar un gráfico que permita verificar si el cambio en el monto de transacciones (cuarto trimestre a primer trimestre) se ajusta a una distribución normal.

```python
# YOUR CODE HERE
import plotly.figure_factory as ff

fig = ff.create_distplot([df['Total_Amt_Chng_Q4_Q1']], group_labels = ['Total_Amt_Chng_Q4_Q1'], bin_size=.1,
                         curve_type='normal'
                        )

fig.update_layout(title_text='Normal Distribution')
fig.show()
```

### Ejercicio 4
Utiliza `ipywidgets` para filtrar las observaciones de la base de datos de acuerdo con los siguentes criterios:
1. Nivel máximo de escolaridad alcanzado.
2. Número de niños en el hogar.
3. Número de adolescentes en el hogar.

```python
def filtrar(Education_criterio, Kidhome_criterio, Teenhome_criterio):
    # Filtra el dataframe acorde a los criterios escogidos
    if all(criterio == "Todos" for criterio in (Education_criterio, Kidhome_criterio, Teenhome_criterio)) is False:
        display(mercadeo[(mercadeo["Education"] == Education_criterio if Education_criterio != "Todos" else True) &
                         (mercadeo["Kidhome"] == Kidhome_criterio if Kidhome_criterio != "Todos" else True) &
                         (mercadeo["Teenhome"] == Teenhome_criterio if Teenhome_criterio != "Todos" else True)])           
    else:
        display(mercadeo)

# Genera el listado de opciones de una categoría
def actualizar_categorias(columna):
    opciones_categoria = mercadeo[columna].unique().tolist()
    opciones_categoria.sort()
    opciones_categoria.insert(0, "Todos")
    return opciones_categoria

    
interactive_plot = interactive(
                                filtrar, #Filtrar y mostrar el dataframe
                                Education_criterio = actualizar_categorias("Education"), # Dropdown para criterio Education
                                Kidhome_criterio = actualizar_categorias("Kidhome"), # Dropdown para criterio Kidhome
                                Teenhome_criterio = actualizar_categorias("Teenhome") # Dropdown para criterio Teenhome
                              )

display(interactive_plot)
```

### Ejercicio 5
Utiliza `ipywidgets` y `plotly` para generar un gráfico con controles que:
1. Cuente con un control para seleccionar una de las variable de Gastos frente a los ingresos.
1. Genere un gráfico de dispersión
2. Genere un diagrama de caja.

```python
gasto = widgets.Dropdown(
                        description = 'Variable de gastos:',
                        options = ["MntWines", "MntFruits", "MntMeatProducts",
                                   "MntFishProducts", "MntSweetProducts", "MntGoldProds"],
                        style = {'description_width': 'initial'}       # Mostrar el nombre completo.
                       )

# Diagrama de dispersión
fig_scatter = go.FigureWidget(
    data = go.Scatter(mode = 'markers', y = mercadeo["Income"]),
    layout = go.Layout(
        xaxis_title = 'Gasto',
        yaxis_title = 'Income',
        title_text = 'Dispersión Gasto {} vs. Ingreso'.format(gasto.value)))

# Diagrama de caja
fig_box = go.FigureWidget(
    data = go.Box(y = mercadeo["Income"], name = "Ingreso"),
    layout = go.Layout(
        title = 'Diagrama de caja: Ingreso')
)

# Filtro de los datos acorde al tipo de gasto escogido
def filtrar():
    filtrado = mercadeo[[gasto.value, "Income"]]
    # Cambiar los datos de X para el diagrama de dispersión según el tipo de gasto elegido
    fig_scatter.data[0].x = filtrado[gasto.value]
    fig_scatter.update_layout(title_text = 'Dispersión Gasto {} vs. Ingreso'.format(gasto.value))

# Filtrar los datos cuando el usuario cambie la opción de la lista despegable
def respuesta(change):
    filtrar()

# Observar el evento de cambio de opción de la lista despegable
gasto.observe(respuesta, names = 'value')

# Maquetación de la visualización
parte_superior = widgets.HBox([gasto]) # # Orden horizontal de lista despegable
parte_inferior = widgets.HBox([fig_scatter, fig_box]) # Orden horizontal de gráficos
visualizacion = widgets.VBox([parte_superior, parte_inferior]) # Orden vertical de ambos elementos

display(visualizacion)
```
